```python
import numpy as np
import pandas as pd
df=pd.read_csv("D:\iris.txt",names=[0,1,2,3,4])
df
```

```python
print("shape is:",df.shape)
print("dimn is:",df.ndim)
print("size is:",df.size)
```

```python
ir1,ir2,ir3=np.split(df,3)
print("iris1 is:\n",ir1)
print("iris2 is:\n",ir2)
print("iris3 is:\n",ir3)
```

```python
header=["sepal length","sepal width","petal length","petal width","Species No"]
header
```

```python
iris_max=np.max(df,axis=0)
print("max of columns:\n",iris_max)
```

```python
iris_min=np.min(df,axis=0)
print("min of columns:\n",iris_min)
```

```python
iris_mean=np.mean(df,axis=0)
print("mean of columns:\n",iris_mean)
print(np.round(iris_mean,2))
```

```python
iris_std=np.std(df,axis=0)
print("standard deviation  of columns:\n",iris_std)
print(np.round(iris_std,2))
```

```python
iris_var=np.var(df,axis=0)
print("variance  of columns:\n",iris_var)
print(np.round(iris_var,2))
```
